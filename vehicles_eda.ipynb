```python
import pandas as pd
df = pd.read_json("vehicles_catalog.json")
df["id"].count()
```

### Pre-Process Data

```python
df["price_int"] = df["price_obj"].apply(lambda x: x.get("value"))
# Price Object has the following 'value', 'view', 'period', 'bulk', 'type'

# Extract values into separate columns
df['Condition'] = df['attrs'].apply(
    lambda x: next(
        (item.get("value") for item in x if item.get("name") == 'Condition'), None)
)
df['Transmission'] = df['attrs'].apply(
    lambda x: next(
        (item.get('value') for item in x if item.get('name') == 'Transmission'), None)
)
df['Mileage'] = df['attrs'].apply(
    lambda x: next(
        (f"{item.get('value')} {item.get('unit')}" for item in x if item.get('name') == 'Mileage'), None)
)
```
