```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_json("dataset/property_catalog.json")
df["id"].count()
```

### Pre-Process Data

```python
df["price_int"] = df["price_obj"].apply(lambda x: x.get("value"))
```

```python
short_let_df = df[df["category_id"] == 45].copy()
short_let_df["price_int_log"] = np.log1p(short_let_df["price_int"])

plt.subplot(121)
sns.histplot(short_let_df["price_int"])

# short_let_df["price_int"].skew()
plt.subplot(122)
sns.histplot(short_let_df["price_int_log"])
```
